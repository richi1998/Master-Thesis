```python
import pandas as pd
import numpy as np
import csv
import ast
```

```python
lingModRes = pd.read_csv('semModRes.csv', encoding="utf-8")
```

```python
bertDF = pd.read_csv('bertresult.csv', encoding="utf-8")
```

```python
bertDF
```

```python
# Function to extract label and score
def extract_label_score(row):
    result = ast.literal_eval(row['sentence'])
    return result[0]['label'], result[0]['score']

# Apply function to extract label and score
bertDF[['label', 'score']] = bertDF.apply(extract_label_score, axis=1, result_type='expand')

# Display the dataframe
print(bertDF[['label', 'score']])
```

```python
lingModRes['paragrafnummer']
```

```python
# Get unique labels from bertDF
unique_labels = bertDF['label'].unique()

# Iterate over unique labels
for label in unique_labels:
    # Get the indices where label matches in bertDF
    label_indices = bertDF.index[bertDF['label'] == label].tolist()
    
    # Iterate over indices and map label and score to lingModRes based on 'paragrafnummer'
    for idx in label_indices:
        paragrafnummer = idx + 1  # Assuming paragrafnummer starts from 1
        label_value = label
        score_value = bertDF.loc[idx, 'score']
        lingModRes.loc[lingModRes['paragrafnummer'] == paragrafnummer, 'label'] = label_value
        lingModRes.loc[lingModRes['paragrafnummer'] == paragrafnummer, 'score'] = score_value
        
# Display lingModRes to verify the addition
lingModRes
```
